### .drop()
##### DataFrameのSeriesを落とす

```python
df = pd.read_csv('tmdb/tmdb_5000_movies.csv')
df.head()
```

```python
df.drop(0)
```

```python
df.drop('id', axis=1)
# カラムを落とすときは axis = 1を指定する
```

```python
df
# 元のdfは変更されていない
```

```python
df.drop('id', axis=1, inplace = True)
df
# inplace = Trueでdfを落として置き換える
```

```python
df = df.drop(0)
df
#df自体を更新してしまうこともできる
```

# .fillna()
### 欠損値を埋める

```python
df['runtime'].mean()
# .mean()でリストの平均値を返す
```

```python
df['runtime'].fillna(df['runtime'].mean())
```

```python
# dfの再定義で欠損値代入したシリーズで置換
df['runtime'] = df['runtime'].fillna(df['runtime'].mean())
```

```python
df[np.isnan(df['runtime'])]
# np.isnanでNaNのシリーズがあるかを確認
```

## .isna()

```python
df = pd.read_csv('tmdb/tmdb_5000_movies.csv')
df.isna()
# NaN = True, !NaN = False が返ってくる
```

```python
df['homepage'].isna()
# df[np.isnan(df['homepage'])] と同じ
```

```python
df[df['homepage'].isna()]
```

```python
# 新しいカラムの作成
df['homepage_nan'] = df['homepage'].isna()
df
```

```python
df[df['homepage_nan']]
# 'homepage_nan' = True を返す
```

```python
pd.isna(df['homepage'])
```
