```python
# Settings
start = '2019-06-01'
end = '2020-05-01'
company_code = '6502.JP'
```

```python
# 
df = data.DataReader(company_code, 'stooq')
df = df[(df.index>=start) & (df.index<=end)]

date = df.index
price = df['Close']

span5 = 5
span25 = 25
span50 = 50

df['sma5'] = price.rolling(window=span5).mean()
df['sma25'] = price.rolling(window=span25).mean()
df['sma50'] = price.rolling(window=span50).mean()

plt.figure(figsize=(30,15))
plt.subplot(2,1,1)

plt.plot(date,price,label='Close', color='#99b898')
plt.plot(date,df['sma5'],label='sma5', color='#e84a5f')
plt.plot(date,df['sma25'],label='sma25', color='#ff847c')
plt.plot(date,df['sma50'],label='sma50', color='#feceab')
plt.legend()

plt.subplot(2,1,2)
plt.bar(date,df['Volume'], label = 'Volume', color='grey')
plt.legend()
```

```python
# functionalization
def company_stock(start, end, company_code):
    df = data.DataReader(company_code, 'stooq')
    df = df[(df.index>=start) & (df.index<=end)]

    date = df.index
    price = df['Close']

    span5 = 5
    span25 = 25
    span50 = 50

    df['sma5'] = price.rolling(window=span5).mean()
    df['sma25'] = price.rolling(window=span25).mean()
    df['sma50'] = price.rolling(window=span50).mean()

    plt.figure(figsize=(30,15))
    plt.subplot(2,1,1)

    plt.plot(date,price,label='Close', color='#99b898')
    plt.plot(date,df['sma5'],label='sma5', color='#e84a5f')
    plt.plot(date,df['sma25'],label='sma25', color='#ff847c')
    plt.plot(date,df['sma50'],label='sma50', color='#feceab')
    plt.legend()

    plt.subplot(2,1,2)
    plt.bar(date,df['Volume'], label = 'Volume', color='grey')
    plt.legend()
```

```python
company_stock('2019-06-01', '2020-06-01', '6502.JP')
```

```python
company_stock('2017-06-01', '2020-06-01', '6502.JP')
```
